```python
# BONUS: Put all the previously mentioned data transformations into a function.

def month(df):
    
    cols = []
    for column in df.columns:
        cols.append(column.lower())
    df.columns = cols
    
    df = df.rename(columns={'unnamed: 0':'index'})
    df = df.drop(['index'], axis=1)
    
    df['response'] = df['response'].fillna('NA')
    df['state'] = df['state'].fillna('NA')
    df['vehicle class'] = df['vehicle class'].fillna('NA')
    df['vehicle size'] = df['vehicle size'].fillna('NA')
    
    df = df.dropna(subset=['months since last claim'])
    df = df.drop(['vehicle type'], axis=1)
    
    df['date'] = pd.to_datetime(df['effective to date'])
    df['month'] = df['date'].dt.month
    filtered = df[df['month']<= 3]
    
    return filtered
```

```python
#copying the original dataframe to a new variable to see if it works
data3 = pd.read_csv("marketing_customer_analysis.csv")
```

```python
#It worked!
month(data3)
```
